```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk,spacy, string
import re
import en_core_web_sm

nlp = en_core_web_sm.load()
```

```python
# Function to lean clean data
def clean_text(text):
    text = text.lower()  # Make the text lowercase
    text = re.sub('\[.*\]','', text).strip() # Remove text in square brackets
    text = text.translate(str.maketrans('', '', string.punctuation)) # Remove punctuation
    text = re.sub('\S*\d\S*\s*','', text).strip()  # Remove words containing numbers
    return text.strip()

stopwords = nlp.Defaults.stop_words

def lemmatizer(text):
    doc = nlp(text)
    sent = [token.lemma_ for token in doc if not token.text in set(stopwords)]
    return ' '.join(sent)
```

Now we will make a copy of the data frame and apply these functions. We're going to make predictions based on the descriptions.

```python
clean_data = news_catagories_data.copy()

clean_data['short_description'] = news_catagories_data['short_description'].apply(clean_text)
```

```python
clean_data['lemma'] = news_catagories_data['short_description'].apply(lemmatizer)
```

```python
clean_data = clean_data[[ 'short_description', 'lemma', 'category']]
```

We have to now extract POS tags

```python
def extract_pos_tags(text):
    doc = nlp(text)
    sent = [token.text for token in doc if token.tag_ == 'NN']
    return ' '.join(sent)

clean_data['short_description_POS_removed'] =  clean_data.lemma.apply(lambda x: extract_pos_tags(x))
```

```python
clean_data['description_Clean'] = clean_data['short_description_POS_removed'].str.replace('-PRON-', '')
clean_data = clean_data.drop(['short_description_POS_removed'],axis = 1)
```

Now we can preform our feature extraction and vectorize our data.


```python
tfidf = TfidfVectorizer(min_df=2, max_df=0.95, stop_words='english')

dtm =  tfidf.fit_transform(clean_data.description_Clean)

len(tfidf.get_feature_names_out())
```

Now we have everything prepared to build our model. We first have to preform our topic modeling. 

```python
clean_data['category'].unique()
```

```python
from sklearn.decomposition import NMF

num_topics =  42

nmf_model = NMF(n_components=num_topics, random_state=27)
nmf_model_2 = NMF(n_components=num_topics, random_state=27)

W1 = nmf_model.fit_transform(dtm)
H1 = nmf_model.components_

```

```python
num_words=30

vocab = np.array(tfidf.get_feature_names_out())

top_words = lambda t: [vocab[i] for i in np.argsort(t)[:-num_words-1:-1]]
topic_words = ([top_words(t) for t in H1])
topics = [' '.join(t) for t in topic_words]
```

```python
colnames = ["Topic" + str(i) for i in range(nmf_model.n_components)]
docnames = ["Doc" + str(i) for i in range(len(clean_data.description_Clean))]
df_doc_topic = pd.DataFrame(np.round(W1, 2), columns=colnames, index=docnames)
significant_topic = np.argmax(df_doc_topic.values, axis=1)
df_doc_topic['dominant_topic'] = significant_topic

clean_data['Topic'] = significant_topic


```

Now we can finally build our models, we will try a couple out and see which works best.

```python
model_data = clean_data[['short_description','Topic']]

# reverse_topic_mapping = {
#     0: 'WORLD NEWS',
#     1: 'TECH',
#     2: 'BUSINESS',
#     3: 'U.S. NEWS',
# }

model_data.head()
```

Now we split the data into our response and predictors and training and test sets,

```python
from sklearn.feature_extraction.text import TfidfTransformer

X = model_data.short_description
y = model_data.Topic

count_vect = CountVectorizer()
X_vect = count_vect.fit_transform(X)

tfidf_transformer = TfidfTransformer()
X_tfidf = tfidf_transformer.fit_transform(X_vect)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.25, random_state=27, stratify=y)
```

We'll now create and test our models

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_recall_fscore_support

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

def display_classification_report(model,metric):
    
    y_train_pred_proba = model.predict_proba(X_train)
    y_test_pred_proba = model.predict_proba(X_test)
    roc_auc_score_train = round(roc_auc_score(y_train, y_train_pred_proba,average='weighted',multi_class='ovr'),2)
    roc_auc_score_test = round(roc_auc_score(y_test, y_test_pred_proba,average='weighted',multi_class='ovr'),2)
    print("ROC AUC Score Train:", roc_auc_score_train)
    print("ROC AUC Score Test:", roc_auc_score_test)
    metric.append(roc_auc_score_train)
    metric.append(roc_auc_score_test)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
  
    precision_train,recall_train,fscore_train,support_train=precision_recall_fscore_support(y_train,y_train_pred,average='weighted')
    precision_test,recall_test,fscore_test,support_test=precision_recall_fscore_support(y_test,y_test_pred,average='weighted')
    
    acc_score_train = round(accuracy_score(y_train,y_train_pred),2)
    acc_score_test = round(accuracy_score(y_test,y_test_pred),2)
    
    metric.append(acc_score_train)
    metric.append(acc_score_test)
    metric.append(round(precision_train,2))
    metric.append(round(precision_test,2))
    metric.append(round(recall_train,2))
    metric.append(round(recall_test,2))
    metric.append(round(fscore_train,2))
    metric.append(round(fscore_test,2))
    
    print('Train Accuracy :',acc_score_train)
    print('Test Accuracy :',acc_score_test)
        
    model_report_train = classification_report(y_train,y_train_pred)
    model_report_test = classification_report(y_test,y_test_pred)
    
    print('Classification Report for Train:\n',model_report_train)
    print('Classification Report for Test:\n',model_report_test)


    # Plot the confusion matrix
    fig, ax = plt.subplots(figsize=(12, 8))
    # Create the confusion matrix
    cm = confusion_matrix(y_test, y_test_pred)
    cmp = ConfusionMatrixDisplay(cm, display_labels=model.classes_)
    cmp.plot(ax=ax)
    plt.xticks(rotation=80)

    plt.show()
```

```python
folds = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 27)

# Cross validation 

def grid_search(model,folds,params,scoring):
    
    grid_search = GridSearchCV(model,
                                cv=folds, 
                                param_grid=params, 
                                scoring=scoring, 
                                n_jobs=-1, verbose=1)
    return grid_search

def print_best_score_params(model):
    print("Best Score: ", model.best_score_)
    print("Best Hyperparameters: ", model.best_params_)
```

```python
bayes = MultinomialNB()

bayes_params = {  
'alpha': (1, 0.1, 0.01, 0.001, 0.0001)  
}

bayes_grid_search = grid_search(bayes, folds, bayes_params, scoring=None)

bayes_grid_search.fit(X_train,y_train)


display_classification_report(bayes_grid_search,[]) 
```

```python
logit = LogisticRegression(random_state=27, solver='lbfgs',multi_class='multinomial')

logit.fit(X_train,y_train)

# Cross validation was taking too long with logistic regression!

display_classification_report(logit,[])
```

```python
tree = DecisionTreeClassifier(random_state=27)

tree_params = {
    'max_depth': [3, 5, 7, 9, 11, 13, 15],
    'min_samples_split': [2, 4, 6, 8, 10, 12, 14]
}


grid_search_tree = grid_search(tree, folds, tree_params, scoring=None)


grid_search_tree.fit(X_train,y_train)

display_classification_report(grid_search_tree,[])
```

```python
forest = RandomForestClassifier(random_state=27, n_estimators=1000, n_jobs=-1,oob_score=True)

forest_params = {
    'max_depth': [3, 5, 7, 9, 11, 13, 15],
    'min_samples_split': [2, 4, 6, 8, 10, 12, 14]
}

grid_search_forest = grid_search(forest, folds, forest_params, scoring=None)

grid_search_forest.fit(X_train,y_train)


display_classification_report(grid_search_forest,[])
```

We can see that by far the logistic regression model worked the best so we will be using that one.
