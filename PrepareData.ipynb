So now we have only Business news aticles, we can also look to seperate into US and global news. We want any mention of US cities, states, etc to be classed as US news, and everything else to be world. 

```python


pd.set_option('display.max_rows', 200)
articles.tail(500)
```

We can see that there are many different headlines, by fortunately they all follow a similar format. It appears that the only American headlines are ones that include "Washington" or "New York", so we will do some simple pattern matching to find those. 

```python
import re 

american_locations_pattern = r'\b(?:new york|washington|newyork)\b' 


def check_us_location(text):
    return bool(re.search(american_locations_pattern, text.lower()))


articles['Location'] = articles['Article'].apply(lambda x: 'US' if check_us_location(x) else 'Global')
```

```python

articles
```

We can see that this didn't work perfectly, but it is okay since some were classes as US because they included New York or Washington in the headline so it should be close enough. We can split the data set into US and Global articles now as well.

```python
us_articles = articles[articles['Location'] == 'US'].reset_index()
global_articles = articles[articles['Location'] == 'Global'].reset_index()
```
