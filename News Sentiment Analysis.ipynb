# Sentiment Analysis

The next blocks of code conducts sentiment analysis on the news headlines collected. We create 6 dataframe of news from different sources and we get the subjectivity and polarity of the headlines.

We first load the datasets we have processed below into pandas dataframes.

```python
import pandas as pd
```

```python
business = pd.read_csv("business_news.csv")
us_news = pd.read_csv("us_news.csv")
tech = pd.read_csv("tech_news.csv")
world = pd.read_csv("world_news.csv")
c1 = pd.read_csv("commodities1.csv")
c2 = pd.read_csv("commodities2.csv")
```

Now, we clean the data. We first drop the columns that are unnecessary for the analysis. Then, we drop the rows with missing data that we need.

```python
business = business.drop(columns = ['Unnamed: 0'])
us_news = us_news.drop(columns = ['Unnamed: 0'])
tech = tech.drop(columns = ['Unnamed: 0', 'index'])
world = world.drop(columns = ['Unnamed: 0'])
c1 = c1.drop(columns = ['author'])
c2 = c2.drop(columns = ['author'])


business = business.dropna().copy()
us_news = us_news.dropna().copy()
tech = tech.dropna().copy()
world = world.dropna().copy()
c1 = c1.dropna().copy()
c2 = c2.dropna().copy()
```

Now, we conduct sentiment analysis and find the subjectivity and polarity of all the news headlines and the article content and we save them in the dataframes as separate columns for further analysis. We used the library TextBlob to do the analysis. 

```python
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS

def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

business['headline'] = business['headline'].astype('string')
business['Subjectivity'] = business['headline'].apply(getSubjectivity)
business['Polarity'] = business['headline'].apply(getPolarity)

us_news['Heading'] = us_news['Heading'].astype('string')
us_news['Subjectivity'] = us_news['Heading'].apply(getSubjectivity)
us_news['Polarity'] = us_news['Heading'].apply(getPolarity)

tech['headline'] = tech['headline'].astype('string')
tech['Subjectivity'] = tech['headline'].apply(getSubjectivity)
tech['Polarity'] = tech['headline'].apply(getPolarity)

world['Heading'] = world['Heading'].astype('string')
world['Subjectivity'] = world['Heading'].apply(getSubjectivity)
world['Polarity'] = world['Heading'].apply(getPolarity)

c1['title'] = c1['title'].astype('string')
c1['Subjectivity'] = c1['title'].apply(getSubjectivity)
c1['Polarity'] = c1['title'].apply(getPolarity)

c2['title'] = c2['title'].astype('string')
c2['Subjectivity'] = c2['title'].apply(getSubjectivity)
c2['Polarity'] = c2['title'].apply(getPolarity)

business['short_description'] = business['short_description'].astype('string')
business['Content Subjectivity'] = business['short_description'].apply(getSubjectivity)
business['Content Polarity'] = business['short_description'].apply(getPolarity)

us_news['Article'] = us_news['Article'].astype('string')
us_news['Content Subjectivity'] = us_news['Article'].apply(getSubjectivity)
us_news['Content Polarity'] = us_news['Article'].apply(getPolarity)

tech['short_description'] = tech['short_description'].astype('string')
tech['Content Subjectivity'] = tech['short_description'].apply(getSubjectivity)
tech['Content Polarity'] = tech['short_description'].apply(getPolarity)

world['Article'] = world['Article'].astype('string')
world['Content Subjectivity'] = world['Article'].apply(getSubjectivity)
world['Content Polarity'] = world['Article'].apply(getPolarity)

c1['content'] = c1['content'].astype('string')
c1['Content Subjectivity'] = c1['content'].apply(getSubjectivity)
c1['Content Polarity'] = c1['content'].apply(getPolarity)

c2['content'] = c2['content'].astype('string')
c2['Content Subjectivity'] = c2['content'].apply(getSubjectivity)
c2['Content Polarity'] = c2['content'].apply(getPolarity)
```
